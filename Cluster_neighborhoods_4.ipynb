```python
import pandas as pd
```

Reading HTML page

```python
df=pd.read_html('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M',match='Postal Code')[0]
```

Ignore cells with a borough that is Not assigned

```python
df['Borough'] = df['Borough'].apply(lambda x:x if x!='Not assigned' else None)
```

```python
df=df.dropna()
```

If a cell has a borough but a Not assigned  neighborhood, then the neighborhood will be the same as the borough.

```python
df['Neighbourhood']=df.apply(lambda x:x['Borough'] if x['Neighbourhood']=='Not assigned' else x['Neighbourhood'],axis=1)
```
